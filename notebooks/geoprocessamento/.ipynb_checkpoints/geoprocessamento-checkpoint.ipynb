```python
import geopandas as gpd
import matplotlib
```

```python
path_list = ['municipios.shp','rodovias2024.shp'] 
dados = [gpd.read_file(i) for i in path_list]
```

```python
municipios_filter = dados[0][dados[0]['SIGLA_UF'] == 'RJ']
rodovias = gpd.overlay(dados[1],municipios_filter, how='intersection')
```
